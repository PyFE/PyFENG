## Brief Introduction
### 1.Model Set up
    
&emsp;&emsp;This paper defines the lifted Heston model as a conventional stochastic volatility model, with n (n ∈ N)
factors for the variance process:  

$$dS_t^n = S_t^n \sqrt[]{V_t^n} dB_t, S^n_0 > 0, $$
$$V_t^n = g^n_0(t) + \sum^n_{i=1} c^n_i U^{n,i}_t ,$$
$$dU_t^{n,i} = ( −x^n_i U^{n,i}_ t − λV^n_t ) dt + ν\sqrt[]{V_t^n} dW_t, $$
$$U_0^{n,i} = 0, i = 1, ... , n, $$


&emsp;&emsp;where λ, ν ∈ R+, 

&emsp;&emsp;$c^n_i , x^n_i ≥ 0$, for i =1, ... , n, 

&emsp;&emsp;$B = ρW + \sqrt[]{1 − ρ^2}W^⊥$, with $(W, W^⊥) $ a two dimensional Brownian motion with ρ ∈ [−1, 1].

&emsp;&emsp;$g^n_0$ is a continuous function on t

##### &emsp;&emsp;[Conditions for $g^n_0$]
&emsp;&emsp;if $g^n_0$ is non-decreasing such that $g^n_0(0) ≥ 0$, or

&emsp;&emsp;$g^n_0 : t → V_0 + \sum_{i=1}^{n}c^n_i \int_0^t e^{−x^n_i (t−s)}θ (s) ds$, with  $V_0$, θ ≥ 0,

&emsp;&emsp;then the strong solution of $(S^n, V^n,(U^{n,i})_{1≤i≤n})$ exists, such that $V_t^n$ ≥ 0, for all t ≥ 0,(proof in Appendix1)

&emsp;&emsp;Specifically, let
$g^n_0 : t → V_0 + \lambdaθ \sum_{i=1}^{n}c^n_i \int_0^t e^{−x^n_i (t−s)}ds$, with  $V_0$, θ ≥ 0



##### &emsp;&emsp;Setting n=1, $c_1^1$ = 1 and $x_1^1$ = 0, it's the standard Heston model

### 2.Approximation of Rough Heston to reduce parameters
&emsp;&emsp;The weights and mean reversions:

$$c^n_i , x^n_i ≥ 0, $$ for  i =1, ... , n 

&emsp;&emsp;They can be calculated by H and r_n, according to the approximation result (Abi Jaber and El Euch 2019, Theorem 3.5), that is, $c^n_i , x^n_i $ is chosen in such a way that sending the number of factors n → ∞ would yield the convergence of the lifted Heston model towards a rough Heston model with parameters (V0, θ, λ, ν, ρ, H).

&emsp;&emsp;that is, $$\Theta_0:= (V_0, θ, λ, ν, ρ, H)$$
$$ volitility smile: σ_n(K, T; r_n, \Theta_0) → σ_∞(K, T; \Theta_0),$$ for each pair of (K,T)

$$c^n_i = \cfrac{(r^{1−α}_n − 1)r_n^{(α−1)(1+n/2) }}{\Gamma(α)\Gamma(2 − α)} r^{(1−α)i}_ n $$

$$x^n_i = \cfrac{1 − α}{ 2 − α} \cfrac{r^{2−α}_n − 1}{r^{1−α}_n−1} r^{i−1−n/2}_n ,  i = 1, ... , n, $$
&emsp;&emsp;where α := H + 1/2 for some H ∈ (0, 1/2)

&emsp;&emsp;and the sequence $(r_n)_{n≥1}$ satisfies $r_n$ ↓ 1 and $n  ln r_n$ → ∞ as n → ∞


### 3.Tractability
&emsp;&emsp;Like the classical Heston model, the lifted Heston model remains tractable. Specifically, fix u ∈ C such that Re(u) ∈
[0, 1]. The Fourier–Laplace transform of the log-price is exponentially affine with respect to
the factors (Un,i)1≤i≤n:

$$E\left[exp(ulogS^n_t)|Ft\right]= exp \left(φ^n(t, T) + u log S_t^n + \sum_{i=1}^n c^n_iψ^{n,i}(T − t)U_t^{n,i}\right)$$

&emsp;&emsp;for all t ≤ T, where $(ψ^{n,i})_{1≤i≤n}$ solves the following n-dimensional system of Riccati ordinary differential equations:

$$(ψ^{n,i})' = −x^n_iψ^{n,i} + F \left(u, \sum_{j=1}^n c^n_j ψ^{n,i}\right), ψ^{n,i}(0) = 0, i = 1, ... , n $$

with
$$F(u, v) = \cfrac{1}{2}(u^2 − u) + (ρνu − λ)v + \cfrac{ν^2}{2} v^2 $$
and
$$φ^n(t, T) = \int_{0}^{T−t} F \left(u, \sum_{i=1}^n c^n_j ψ^{n,i}(s)\right)g^n_0(T-s)ds  , t ≤ T. $$


&emsp;&emsp;In particular, for t=0, since $U_0^{n,i}$ = 0 for i = 1, ... , n, the unconditional Fourier–Laplace transform:

$$E\left[exp(u log S_t^n)\right] = exp\left( u log S^n_0 + \sum^T_0 F \left(u, \sum_{i=1}^n ψ^{n,i}(s)\right)g^n_0(T-s)ds \right) $$


##### &emsp;&emsp;Option prices can be calculated:

&emsp;&emsp;(1)Solve the n-dimensional Riccati system of $(ψ^{n,i})'$ using explicit–implicit discretization scheme

$$ψ^{n,i}_0=0,  ψ^{n,i}_{t_{k+1}}=\cfrac{1}{1 + x^n_i\Delta t} \left( ψ^{n,i}_{t_k}+\Delta t F\left( u, \sum_{j=1}^nψ^n_jψ^{n,j}_{t_k} \right) \right),$$ 

$$i = 1, ... , n$$

&emsp;&emsp;(2)Get the characteristic function

&emsp;&emsp;(3)Compute the call prices via the cosine method  for the inversion of the characteristic function(Fang and Oosterlee 2008) 

### 4.Simulation
&emsp;&emsp;Because the Lifted Heston Model is a Markovian and semimartingale model, one can adapt standard recursive Euler–Maruyama schemes,

&emsp;&emsp;this paper  consider the modified explicit–implicit scheme

$$\hat{V^n_{t_k}}= g^n_0(t_k) + \sum^n_{i=1} c^n_i \hat{U^{n,i}_{t_k}} ,   \hat{U_0^{n,i} }= 0 $$

$$\hat{U^{n,i}_{t_{k+1}}} = \cfrac{1}{1 + x^n_i \Delta t } \left(\hat{U^{n,i}_{t_k}} − λ \hat{ V^n_{t_k}} \Delta t + ν\sqrt[]{(\hat V^n_{t_k})^+} (W_{t_{k+1}} − W_{t_k}) \right)	, i = 1, ... , n, $$

where $t_k = k\Delta t, k = 1 ...N, \Delta t = T/N $  and  
$W_{t_{k+1}} − W_{t_k}∼ N (0, \Delta t)=Z_1$

##### [the reason behind using explicit-implicit scheme]

&emsp;&emsp;If using explicit scheme,

$$\hat{U^{n,i}_{t_{k+1}}}=（1-x^n_i \Delta t) \hat{U^{n,i}_{t_k}}-λ \hat {V^n_{t_k}} \Delta t + ν\sqrt[]{(\hat V^n_{t_k})^+} (W_{t_{k+1}} − W_{t_k}) $$

&emsp;&emsp;$x^n_n$ grows very large as n increases. For instance,for n = 20, r20 = 2.5 and H = 0.1, $x^n_n$ = 6417.74. Consequently, if one needs to ensure the stability of the explicit scheme, one needs a very large number of time steps N. In contrast, the implicit Euler
scheme is stable for any number of time steps N.



&emsp;&emsp;Given the paths of $\hat{V^n_{t_k}}$, we can get the pahts of $S^n_{t_k}$, 

$$S^n_{t_{k+1}}=S^n_{t_k}+S^n_{t_k} \sqrt[]{V^n_{t_k}}(B_{t_{k+1}}-B_{t_k})$$

&emsp;&emsp;where $B_{t_{k+1}} − B_{t_k}∼ N (0, \Delta t)=Z_2,  cor(Z_1，Z_2）=\rho $ and finally get $S^n_T$ to calculate the price of options

&emsp;&emsp;This paper compares the estimated Hurst index of realized volatility of S&P Index, Heston Model and Lifted Heston Model,
founding that the Lifted Heston Model enjoys the similar roughness with realized value.

### 5.Strengths and Weakness
##### &emsp;&emsp;Strengths:
&emsp;&emsp;(1)Lifted Heston Model is an approximation to Rough Heston Model, keeping the characteristics of well reproduce the at-the-money skew of the implied volatility surface, therefore catching the roughness of volatility change.

&emsp;&emsp;(2)By introducing n factors, Lifted Heston Model can get a semimartingale of $V_t$, therefore, $S_t$ is tractable and can be more easily simulated to price options.

&emsp;&emsp;(3)Compared with other methods like adding jumps (Cont and Tankov 2003, Gatheral 2011) and stacking additional random factors (Bergomi 2005, Fouque et al. 2011), with more parameters slowing down the calibration process. Lifted Heston Model only introduces one more parameter H than Heston Model.
##### &emsp;&emsp;Weakness:

&emsp;&emsp;(1)Compared with Heston Model, the prices and volatilities can only be simulated through conditional Monte Carlo, simulation process is slow

## Comparison Between Lifted Heston Model and Rough Heston Model

Lifted Heston model nests as extreme cases the classical Heston model (when $n=1$), and the rough Heston model (when $n$ goes to infinity).

In order to visualize the convergence to rough Heston model when $n$ goes to infinity, we generate our benchmark implied volatility surface, for $9$ maturities $T \in \{1w,1m,2m,3m,6m,9m,1yr,1.5yr,2yr\} $, and with $26$ strikes $K$ per maturity.

The parameters of rough Heston model is given by
$$V_0=0.02,\; \theta=0.02,\; \lambda=0.3,\; \nu=0.3,\; \rho=-0.7,\; H=0.1$$

The implied volatiliy surface can be computed by Fourier inversion techniques. For simplicity, we only consider the call option situation.

We first define:
$$x=log(S_0/K),\quad y=log(S_T/K)$$

Then the price of the call European option will be
$$v(y,T)=K(e^y-1)^{+}$$

Using the method of inverse Fourier integral via cosine expansion, we have:
$$price\;of\;call\;option=v(x,T,V_0)\approx Ke^{-r\Delta t}\cdot Re \left\{ \dfrac 1 2 U_0+\sum_{k=1}^{N-1} \phi_{hes}(\dfrac {k\pi}{b-a};V_0)e^{ik\pi\frac {x-a}{b-a}}U_k\right\}$$

where, $\phi_{hes}(\omega;V_0)$ is the characteristic function of the log-asset price $y$, when $x=0$, $[a,b]$ should cover $[x-10\sigma,x+10\sigma]$ (or other sufficiently large interval), with $\sigma$ denoting the standard deviation of the density $f(y|x)$.

$U_k$ is given by:

$$U_k=\frac 2 {b-a}(\chi_k(0,b)-\psi_k(0,b))$$

$$\chi_k(c,d)=\frac 1 {1+(\frac {k\pi}{b-a})^2 }[cos(k\pi\frac {d-a}{b-a})e^d-cos(k\pi \frac{c-a}{b-a})e^c+\frac {k\pi}{b-a} sin(k\pi\frac{d-a}{b-a})e^d-\frac {k\pi}{b-a} sin(k\pi \frac {c-a}{b-a})e^c]$$

$$\psi_k(c,d)=[sin(k\pi\frac {d-a}{b-a})-sin(k\pi \frac {c-a}{b-a})]\frac {b-a}{k\pi},\quad for\; k\neq 0$$
$$\psi_0(c,d)=d-c,\quad for \; k=0$$

$\phi_{hes} (\omega;V_0)$ is derived from the fact that:

$$\mathbb{E}[ {exp}(u{ log} S_T)]= {exp}(u {log} S_0+\int_0^T F(u,\psi(s,u))g_0(T-s)ds)$$

$$F(u,v)=\frac 1 2 (u^2-u)+(\rho \nu u-\lambda)v+\frac {\nu^2}{2}v^2$$

$$g_0(t)=V_0+\lambda\theta\int_0^t \frac {s^{H-1/2}}{\Gamma(H+1/2)}ds$$

$$\psi(t,u)=\frac 1 {\Gamma(H+1/2)}\int_0^t (t-s)^{H-1/2}F(u,\psi(s,u))ds$$

$\psi$ solves the fractional Riccati equation. We can use the Adams Predictor-Corrector scheme for the discretization of the fractional Riccati equation. The procedure is shown below:

$$\psi(t,a)=\dfrac 1 {\Gamma(\alpha)}\int_0^t (t-s)^{\alpha-1}F(a,\psi(s,a))ds$$

Over a regular discrete time-grid with mesh $\Delta$, $0\leq t_0\leq...\leq t_n\leq t$,

$$\psi(t_{k+1},a)\approx \hat \psi (t_{k+1},a)=\sum_{0\leq j\leq k} a_{j,k+1}F(a,\hat \psi(t_j,a))+a_{k+1,k+1}F(a,\hat \psi^P(t_{k+1},a))$$

with
$$a_{0,k+1}=\dfrac {\Delta^\alpha}{\Gamma(\alpha+2)}[k^{\alpha+1}-(k-\alpha)(k+1)^\alpha]$$
$$a_{j,k+1}=\dfrac {\Delta^\alpha}{\Gamma(\alpha+2)}[(k-j+2)^{\alpha+1}+(k-j)^{\alpha+1}-2(k-j+1)^{\alpha+1}]\quad 1\leq j \leq k$$
$$a_{k+1,k+1}=\dfrac {(\Delta t)^\alpha} {\Gamma(\alpha+2)}$$

$$b_{j,k+1}=\dfrac {\Delta^\alpha}{\Gamma(\alpha+1)}((k-j+1)^\alpha-(k-j)^\alpha) $$
$$\hat \psi^P(t_{k+1},a)=\sum_{0\leq j\leq k} b_{j,k+1}F(a,\hat \psi(t_j,a))$$

Following above steps, we can derive the price of the call option under rough Heston model. Since we know the call option price formula of BSM model, we can derive the implied volatility surface through Newton's method with good initial guess.

Call option price formula of BSM model:

$$C=S_0N(d_1)-e^{-rT}KN(d_2)$$
$$d_{1,2}=\dfrac {log(S_0e^{rT}/K)}{\sigma \sqrt{T}}\pm\dfrac 1 2 \sigma \sqrt{T}$$

### Calculation of implied volatility surface of rough Hestion model

Since the calculation speed of Python is relatively slow, we can only choose to run 26 maturies(and this will take many hours). Meanwhile, the time steps of Adams Predictor-Corrector scheme is set at 60 (200 in the paper). The N of cosine method is set at 160, same as in the paper. However, to reduce running time, we will stop the calculation once the variation of price in the last 5 steps is less than $0.005\% S_0$.

```python
def g0(t,V0,λ,θ,H,gamma_coeff):
    ans=V0+λ*θ/gamma_coeff/(H+0.5)*pow(t,H+0.5)
    return ans

def F(u,v,ρ,λ,ν):
    ans=0.5*(u*u-u)+(u*ρ*ν-λ)*v+0.5*ν*ν*v*v
    return ans

#Adams Predictor-Corrector scheme
def ψ(t,u,H,ρ,λ,ν,steps):
    α=H+1/2
    Δ=t/steps
    pow_temp1=pow(Δ,α)
    gamma_coeff2=gamma(α+2)
    gamma_coeff1=gamma_coeff2/(α+1) #gamma(α+1)
    ψ_series=np.zeros(steps+1)
    ψ_series=np.array(ψ_series,dtype=complex)
    ψ_P=np.zeros(steps+1)
    ψ_P=np.array(ψ_P,dtype=complex)
    for k in range(steps):
        a=np.zeros(k+2)
        b=np.zeros(k+1)
        for j in range(k+1):
            a[j]=pow_temp1/gamma_coeff2*(pow(k-j+2,α+1)+pow(k-j,α+1)-2*pow(k-j+1,α+1))
            b[j]=pow_temp1/gamma_coeff1*(pow(k-j+1,α)-pow(k-j,α))
        a[0]=pow_temp1/gamma_coeff2*(pow(k,α+1)-(k-α)*pow(k+1,α))
        a[k+1]=pow_temp1/gamma_coeff2
        for j in range(k+1):
            F_value=F(u,ψ_series[j],ρ,λ,ν)
            ψ_P[k+1]+=b[j]*F_value
            ψ_series[k+1]+=a[j]*F_value
        ψ_series[k+1]+=a[k+1]*F(u,ψ_P[k+1],ρ,λ,ν)
    return ψ_series[steps]

def characteristic_func(u,T,H,ρ,λ,ν,V0,θ,gamma_coeff,steps,intervals):
    ds=T/intervals
    integral_value=0
    ψ_left=ψ(0,u,H,ρ,λ,ν,steps)
    for i in range(intervals):
        left_s=ds*i
        right_s=left_s+ds
        ψ_right=ψ(right_s,u,H,ρ,λ,ν,steps)
        integral_left=F(u,ψ_right,ρ,λ,ν)*g0(T-left_s,V0,λ,θ,H,gamma_coeff)
        integral_right=F(u,ψ_right,ρ,λ,ν)*g0(T-right_s,V0,λ,θ,H,gamma_coeff)
        integral_value+=(integral_left+integral_right)*ds/2
        ψ_left=ψ_right
    ans=np.exp(integral_value)
    return ans

def χ_k(k,c,d,a,b):
    sincos_1=k*np.pi*(d-a)/(b-a)
    sincos_2=k*np.pi*(c-a)/(b-a)
    ans=np.cos(sincos_1)*np.exp(d)-np.cos(sincos_2)*np.exp(c)
    ans+=k*np.pi/(b-a)*(np.sin(sincos_1)*np.exp(d)-np.sin(sincos_2)*np.exp(c))
    ans/=(1+pow(k*np.pi/(b-a),2))
    return ans

def ψ_k(k,c,d,a,b):
    sincos_1=k*np.pi*(d-a)/(b-a)
    sincos_2=k*np.pi*(c-a)/(b-a)
    if (k==0):
        ans=d-c
    else:
        ans=(np.sin(sincos_1)-np.sin(sincos_2))*(b-a)/k/np.pi
    return ans

def U_k(k,a,b,call=1):
    if (call==1):
        ans=(χ_k(k,0,b,a,b)-ψ_k(k,0,b,a,b))*2/(b-a)
    else:
        ans=(-χ_k(k,a,0,a,b)+ψ_k(k,a,0,a,b))*2/(b-a)
    return ans

maturities=[7/365,1/12,1/6,1/4,1/2,3/4,1,1.5,2]
log_moneyness_all=np.zeros(shape=(9,26))
impvol=np.zeros(shape=(9,26))
gamma_coeff=gamma(H+1/2)
for i_T in range(9):
    T=maturities[i_T]
    left_log=-1.2*np.sqrt(T)
    right_log=0.3*np.sqrt(T)
    strikes=np.linspace(left_log,right_log,26)
    for i_strike in range(26):
        log_moneyness=strikes[i_strike]
        log_moneyness_all[i_T,i_strike]=log_moneyness
        x=-log_moneyness
        K=np.exp(-x)*S0
        a=x-1.5
        b=x+1.5
        N=160
        price=0
        check_end=np.zeros(5)
        for i in range(N):
            char_func=characteristic_func(complex(0,i*np.pi/(b-a)),T,H,ρ,λ,ν,V0,θ,gamma_coeff,steps=60,intervals=70)
            price_add=char_func*U_k(i,a,b,1)*np.exp(complex(0,i*np.pi*(x-a)/(b-a)))
            if (i==0):
                price+=price_add.real/2
            else:
                price+=price_add.real
            flag=False
            if (i<5):
                check_end[i]=price*K
            else:
                for i_temp in range(4):
                    check_end[i_temp]=check_end[i_temp+1]
                check_end[4]=price*K
                check_min=check_end[0]
                check_max=check_min
                for i_temp in range(5):
                    check_min=min(check_min,check_end[i_temp])
                    check_max=max(check_max,check_end[i_temp])
                if (check_max-check_min)<0.005*S0/100:
                    flag=True
            if (flag==True):
                break
        price=check_max
      #  print(K,price)
        model=pf.Bsm(0.3)
        impvol[i_T,i_strike]=model._impvol_newton(price,K,S0,T)
```

We can find that, although in our simulation, the results of lifted Heston model and rough Heston model differ a lot, the trend of implied volatility is similar. Meanwhile, the result of rough Heston model is closer to the result in the paper. 

We believe the error in rough Heston model is due to choosing fewer steps in Adams Predictor-Corrector scheme(only 60) and cosine methods(less than 160), as well as fewer intervals to calculate the integral(only 70). 

The error in lifted Hestion model, is probably due to fewer MC simulations(only 1000) and fewer steps when simulating the price and volatility(only 300).
